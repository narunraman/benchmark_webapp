```python
import pandas as pd
import pyvis
```

```python
df = pd.read_pickle('all_tasks.pkl')
df
```

```python
targets = df['dependencies'].apply(lambda x: x[0] if x else x)
sources = df['tasks']
```

```python
task2group_name = {task_name: group_name for group_name, task_name in zip(df['group_name'], df['task_name'])}
```

```python
categories = {
    "Foundations": [
        'Arithmetic', 
        'Probability', 
        'Optimization', 
        'Logic', 
        'Theory of Mind'
    ], 
    "Preferences": [
        'Utility Theory', 
        'Reference Independence'
    ], 
    "Decision Making Under Uncertainty": [
        'Expected Utility Theory', 
        'Risk Assessment'
    ], 
    "Decision Making in the Presence of Other Agents": [
        'Single-Round Game Theory',
        'Multi-Round Game Theory',
    ]
}

rev_categories = {group_name: cat_name for cat_name in categories for group_name in categories[cat_name]}
```

```python
from pyvis.network import Network
from IPython.display import display, HTML

got_net = Network(height="600px", width="1000px", font_color="black", notebook=True, directed = True, cdn_resources="in_line")

# set the physics layout of the network
got_net.barnes_hut()
edge_data = zip(sources, targets, df['group_name'])

colors = ['blue', 'purple', 'green', 'red']
group_color = {name: colors[i] for i, name in enumerate(categories)}

for e in edge_data:
    src = e[0]
    dst = e[1]
    group = e[2]

    got_net.add_node(src, label=src, title=src, borderWidth=2, size = 100, shape='square', group = group)
    if dst:
        for n in dst:
            got_net.add_node(n, label=n, title=n, borderWidth=2, size = 100, shape='square', group = group)
            got_net.add_edge(src, n, width = 30, arrowStrikethrough=False)
neighbor_map = got_net.get_adj_list()

# add neighbor data to node hover data
for node in got_net.nodes:
    node['font'] = {"size": 100}
    # print(node['title'])
    # print(rev_categories)
    node['color'] = {'border': group_color[rev_categories[task2group_name[node['title']]]]}
    # node["title"] += " Neighbors:<br>" + "<br>".join(neighbor_map[node["id"]])


got_net.show_buttons(filter_=True)
got_net.prep_notebook()
got_net.show("dependency.html")
# display(HTML("dependency.html"))
```
